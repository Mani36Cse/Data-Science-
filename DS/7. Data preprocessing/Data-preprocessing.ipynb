```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("chronic_kidney_disease.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
df = data.select_dtypes(include='object')  
# Select object columns
data[df.columns] = df.apply(lambda x: x.str.strip())
data.replace('	?', np.nan, inplace=True)
data.replace('?', np.nan, inplace=True)
```

```python
data['pcv'].unique()
```

```python
#to convert the object data type into numerical data type
data["pcv"]=pd.to_numeric(data["pcv"])
data["wc"]=pd.to_numeric(data["wc"])
data["rc"]=pd.to_numeric(data["rc"])
```

```python
def Univariate(data):
    Qual=[]
    Quan=[]
    for y in data.columns:
        if data[y].dtype=='O':
            Qual.append(y)
        else:
            Quan+=[y]
    return Qual,Quan
```

```python
qual,quan=Univariate(data)
```

```python
#Replace the NaN data into categorical data
```

```python
cat=["pcc","ba","htn","dm","cad","appet","pe","ane","rbc","pc"]
data_qual[cat]=data_qual[cat].fillna(data_qual[cat].mode().iloc[0])
```

```python
data_qual.isnull().sum()
```

```python
nu=["age","bp","sg","bgr","bu","sc","sod","pot","hemo","pcv","wc","rc"]
data_quan[nu]=data_quan[nu].fillna(data_quan[nu].mean())
nm=["al","su"]
data_quan[nm]=data_quan[nm].fillna(data_quan[nm].mode().iloc[0])
```
