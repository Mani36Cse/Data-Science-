```python
df = data.select_dtypes(include='object')  
# Select object columns
data[df.columns] = df.apply(lambda x: x.str.strip())
data.replace('	?', np.nan, inplace=True)
data.replace('?', np.nan, inplace=True)
```

```python
data['pcv'].unique()
```

```python
def Univariate(data):
    Qual=[]
    Quan=[]
    for y in data.columns:
        if data[y].dtype=='O':
            Qual.append(y)
        else:
            Quan+=[y]
    return Qual,Quan
```

```python
qual,quan=Univariate(data)
```
