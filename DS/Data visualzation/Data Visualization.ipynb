### Univariate analysis

```python
import seaborn as sb
```

```python
sb.distplot(dataset["ssc_p"],hist=True,kde=True,kde_kws={'color':'red'})
```

### Bivariate analysis: Distribution plot

```python
sb.jointplot(x="ssc_p",y="salary",data=dataset)
plt.show()
#It will show the Scatter plot and Histogram for X and Y.
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
#Color represents the number of data points within each bin
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
#we can see the more no of curves in max no of repetition in max curve 
```

### Pairplot

```python
sb.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette="husl")
#We give hue="gender", so it was compare the gender in every numerical columns.
```

### Categorical plot

```python
sb.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True)
#we can see how data was splits
```

```python
sb.swarmplot(x="gender",y="ssc_p",data=dataset) 
#It will avoid overlap
```
